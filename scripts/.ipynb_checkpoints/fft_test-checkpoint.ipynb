```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats
import scipy as sp


sns.set_style('darkgrid')
```

```python
#Funci'on para recrear serie de fourier a partir de una selecci'on de frecuencias significativas
def f(Y, x, N):
    total = 0
    for ctr in range(len(Y)):
        total += Y[ctr] * (np.cos(x*ctr*2*np.pi/(N)) + 1j*np.sin(x*ctr*2*np.pi/(N)))
    return np.real(total)
```

```python
#Fast Plot two time series with avaranges 10m 

def two_time_serie(ws1, ws2, name):
    plt.figure(figsize=(30,10))
    
    ws1.plot(label='ws fft', color='darkblue')
    plt.xticks(fontsize=25)
    
    ws2.plot(label='ws 10m', color='red', style='.', markersize=20)
    
    
    plt.xlabel('Tiempo', size=25)
    plt.ylabel('U [$m/s$]', size=25)
    plt.yticks(fontsize=25)
    plt.xticks(fontsize=25)
    plt.legend(fontsize=25)
    
    plt.savefig('../output_figures/fft_test/'+name+'.png', format='png', dpi=360, bbox_inches="tight")
    plt.show()
```

```python
def two_time_serie_fou(ws1, ws2, label1, label2, figname):
    plt.figure(figsize=(30,10))
    
    ws1.plot(label=label1, color='darkblue')
    plt.xticks(fontsize=25)
    
    ws2.plot(label=label2, color='red', style='-', markersize=20)
    
    
    plt.xlabel('Tiempo', size=25)
    plt.ylabel('U [$m/s$]', size=25)
    plt.yticks(fontsize=25)
    plt.xticks(fontsize=25)
    
    plt.legend(fontsize=25)
    plt.savefig(figname+'.png', format='png', dpi=360, bbox_inches="tight")
    plt.show()
```

```python
#ERROR
#ajuste de una distribuci\'on normal (media, std) 

def fitNorm (serie, name):
    plt.figure(figsize=(20,10))
    
    H,bins,bars=plt.hist(serie, bins=40, density=True, color='grey')
    xmin, xmax = plt.xlim()
    xs = np.linspace(xmin,xmax)
    
    
    #parametros (m,std)
    fitNorm.parametros=stats.norm.fit(serie)
    parametros=fitNorm.parametros
    pdf=stats.norm.pdf(xs, parametros[0], parametros[1])
    plt.plot(xs, pdf, 'darkred', linewidth=3, label='Normal')
    plt.plot(parametros[0], stats.norm.pdf(parametros[0], parametros[0], parametros[1]),marker='o', color='orange', alpha=0.6, ms=10, label="mu = %.2f,  std = %.2f" % (parametros[0], parametros[1]))
    
    plt.ylabel('Densidad de Probabilidad', fontsize=18)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.legend(fontsize=18)
    plt.savefig('../output_figures/fft_test/'+name+'.png', format='png', dpi=360, bbox_inches="tight")
    plt.show()
    
    fitNorm.pdf=pdf
    fitNorm.xs=xs
    print(str('mean:')+str(parametros[0]), str('std:')+str(parametros[1]), str('var:')+str(parametros[1]**2))
```

```python
#Import data
df_day=pd.read_csv('../output_data/ventosa_2018_01_20.csv', index_col='Tiempo', parse_dates=True)
df_day_10m=df_day.resample('10T').mean()
```

```python
Ux = df_day.ws[0:3600*24]
N = Ux.size

Y = np.fft.fft(Ux)
X = sp.fftpack.fftfreq(len(Ux), 1/N)


plt.figure(figsize=(20,8))

plt.plot(X[1:N//2], abs(Y.real[1:N // 2]**2))
plt.yscale('log')
plt.xscale('log')


ymin, ymax = plt.ylim()
plt.axvline(x=600, ymin=0, ymax=ymax, ls='--', lw=2, c='r' )
plt.text(x=10**3-300, y=10**-2, s='600Hz', fontsize=15)

plt.axvline(x=60, ymin=0, ymax=ymax, ls='--', lw=2, c='r' )
plt.text(x=10**2-30, y=10**-2, s='60Hz', fontsize=15)

plt.axvline(x=10, ymin=0, ymax=ymax, ls='--', lw=2, c='r' )
plt.text(x=10**1+1, y=10**-2, s='10Hz', fontsize=15)

plt.axvline(x=1, ymin=0, ymax=ymax, ls='--', lw=2, c='r' )

plt.ylabel('Espectro de Potencia [$m^2/s^2$]',fontsize=18)
plt.xlabel('Frecuencia [Hz]', fontsize=18)

plt.savefig('../output_figures/fft_test/PS24h.png', format='png', dpi=360, bbox_inches="tight")
plt.tick_params(labelsize=20)

plt.show()
```

```python
Ux = df_day.ws[0:3600*24]

Y = np.fft.fft(df_day.ws[0:3600*24])
N = len(Y)
print(N)
xs = range(N)


Ux_s= np.asarray([f(Y[0:600], x, N) for x in xs])
m=np.mean(Ux_s)/np.mean(Ux)
Ux_s=Ux_s/m

plt.figure(figsize=(40,10))

plt.plot(xs, Ux, label='Experimental')
plt.plot(xs, Ux_s, color='red', alpha=1, label='Fourier')

    
plt.xlabel('Orden de muestras', size=28)
plt.ylabel('Velocidad de viento [$m/s$]', size=28)
plt.yticks(fontsize=30)
plt.xticks(fontsize=30)
plt.legend(fontsize=25)
plt.savefig('../output_figures/fft_test/fft24h.png', format='png', dpi=360, bbox_inches="tight")


Error = Ux-Ux_s

plt.figure(figsize=(40,10))
plt.plot(xs,Error, label='Error', c='grey')
plt.xlabel('Orden de muestras', size=28)
plt.ylabel('Error [$m/s$]', size=28)
plt.yticks(fontsize=30)
plt.xticks(fontsize=30)
plt.legend(fontsize=25)
plt.savefig('../output_figures/fft_test/erro24h.png', format='png', dpi=360, bbox_inches="tight")


df_error1=pd.DataFrame(Error)
df_error1['Fou']=Ux_s
df_error1['total']=Ux_s+Error
df_error1.to_csv('../output_data/fft_test/error_24h.csv')
```
